```python
#sort data :sor data in ascending or descending order

#group by command:group your data by particular column you like
```

```python
file_path="mtcars.csv"   #practice
cars=pd.read_csv(file_path)
cars.tail()
```

```python
cars.sort_values(by=['hp'],ascending=False)    #sort data
```

```python
cars.sort_values(by = ["mpg"], ascending=False)
```

```python
#assign column names  ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb'] using columns att
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb'] 
cars.head()
```

```python
#grouping
cars_groups = cars.groupby(cars['am'])
cars_groups.mean()
```

```python
cars_grp= cars.groupby(cars['wt'])   #practice   #group by column you like
cars_grp.mean()
```
